# Predictive Analysis and Hyperparameter Tuning

The prediction of fake news mail is a classification problem. Therefore, we can use some popular Machine Learning Classifiers to make the prediction and then compare their results. Also, we are going to use a Neural Network to learn the Word Embeddings from our Model and measure its performance.

Our task at first is to detect the best Hyperparameters for the classifiers, analysze their performance on our dataset, and then select the best Predictive Model. We are going to use **Grid Search** to detect the best Hyperparameters for the classifiers. The classifiers that can be used for this comparison purpose are mentioned below

1. *Decision Tree*
2. *Random Forest*
3. *Logistic Regression*
4. *Multinomial Naive Bayes*
5. *Support Vector Machines with Gaussian Kernel*

```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
from scipy import sparse
from sklearn.model_selection import train_test_split
import pickle as pkl
```

```python
feature_mtx = np.load('model-features/feature_matrix.npy')
X_tfidf_sparse = sparse.load_npz('model-features/tf-idf-matrix.npz')
X_word_embed = np.load('model-features/wordembedding_feature.npy')

X_tfidf = np.concatenate((feature_mtx, X_tfidf_sparse.toarray()), axis=1)
X_embedding = np.concatenate((feature_mtx, X_word_embed), axis=1)

label = np.load('model-features/label.npy')
```

```python
#With Hyper Parameters Tuning
#Randomforest with TF-IDF
#importing modules
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
#making the instance
model=RandomForestClassifier()
#hyper parameters set
params = {'criterion':['gini','entropy'],
          'n_estimators':[10,15,20,25,30],
          'min_samples_leaf':[1,2,3],
          'min_samples_split':[3,4,5,6,7], 
          'random_state':[0],
          'n_jobs':[4]}
#Making models with hyper parameters sets
model1 = GridSearchCV(model, param_grid=params, n_jobs=4)
train_features6, test_features6, train_labels6, test_labels6= train_test_split(X_tfidf, label, test_size=0.3, random_state=0)
#learning
model1.fit(train_features6,train_labels6)
#The best hyper parameters set
print("Best Hyper Parameters:\n",model1.best_params_)
best_paraeter=model1.best_params_
Optimised_LR_TFIDF=model1.best_estimator_
pkl.dump(best_paraeter,open('RF_TF-ID_param.pkl','wb'))
pkl.dump(Optimised_LR_TFIDF,open('RF_TF-ID_clf.pkl','wb'))
```

```python
#With Hyper Parameters Tuning
#Randomforest with skipgram
#importing modules
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
#making the instance
model=RandomForestClassifier()
#hyper parameters set
params = {'criterion':['gini','entropy'],
          'n_estimators':[10,15,20,25,30],
          'min_samples_leaf':[1,2,3],
          'min_samples_split':[3,4,5,6,7], 
          'random_state':[0],
          'n_jobs':[4]}
#Making models with hyper parameters sets
model1 = GridSearchCV(model, param_grid=params, n_jobs=4)
train_features7, test_features7, train_labels7, test_labels7= train_test_split(X_embedding, label, test_size=0.3, random_state=0)
#learning
model1.fit(train_features7,train_labels7)
#The best hyper parameters set
print("Best Hyper Parameters:\n",model1.best_params_)
best_paraeter=model1.best_params_
Optimised_LR_skipgram=model1.best_estimator_
pkl.dump(best_paraeter,open('RF_skipgram_param.pkl','wb'))
pkl.dump(Optimised_LR_skipgram,open('RF_skipgram_clf.pkl','wb'))
```
