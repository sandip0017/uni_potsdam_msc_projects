# Predictive Analysis and Hyperparameter Tuning

The prediction of fake news mail is a classification problem. Therefore, we can use some popular Machine Learning Classifiers to make the prediction and then compare their results. Also, we are going to use a Neural Network to learn the Word Embeddings from our Model and measure its performance.

Our task at first is to detect the best Hyperparameters for the classifiers, analysze their performance on our dataset, and then select the best Predictive Model. We are going to use **Grid Search** to detect the best Hyperparameters for the classifiers. The classifiers that can be used for this comparison purpose are mentioned below

1. *Decision Tree*
2. *Random Forest*
3. *Logistic Regression*
4. *Multinomial Naive Bayes*
5. *Support Vector Machines with Gaussian Kernel*

```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
from scipy import sparse
from sklearn.model_selection import train_test_split
import pickle as pkl
```

```python
feature_mtx = np.load('model-features/feature_matrix.npy')
X_tfidf_sparse = sparse.load_npz('model-features/tf-idf-matrix.npz')
X_word_embed = np.load('model-features/wordembedding_feature.npy')

X_tfidf = np.concatenate((feature_mtx, X_tfidf_sparse.toarray()), axis=1)
#X_Embeding
X_embedding = np.concatenate((feature_mtx, X_word_embed), axis=1)

label = np.load('model-features/label.npy')
```

```python
#With Hyper Parameters Tuning
#SVM with TF-IDF
#importing modules
from sklearn.model_selection import GridSearchCV
from sklearn import svm
from sklearn.model_selection import train_test_split
#making the instance
model=svm.SVC()
#Hyper Parameters Set
params = {'C': np.linspace(1000,2000,10),
          'kernel': ['rbf']}
#Making models with hyper parameters sets
model1 = GridSearchCV(model, param_grid=params, n_jobs=4)
train_features2, test_features2, train_labels2, test_labels2 = train_test_split(X_tfidf, label, test_size=0.3, random_state=0)
#Learning
model1.fit(train_features2,train_labels2)
#The best hyper parameters set
print("Best Hyper Parameters:\n",model1.best_params_)
optimised_SVM_tfidf=model1.best_estimator_
pkl.dump(model1.best_params_,open('SVM_TF-ID_param.pkl','wb'))
pkl.dump(model1.best_estimator_,open('SVM_TF-ID_clf.pkl','wb'))
```

```python
#With Hyper Parameters Tuning
#SVM with skipgram
#importing modules
from sklearn.model_selection import GridSearchCV
from sklearn import svm
from sklearn.model_selection import train_test_split
#making the instance
model=svm.SVC()
#Hyper Parameters Set
params = {'C': np.linspace(1000,2000,10), 
          'kernel': ['rbf']}
#Making models with hyper parameters sets
model1 = GridSearchCV(model, param_grid=params, n_jobs=4)
train_features3, test_features3, train_labels3, test_labels3 = train_test_split(X_embedding, label, test_size=0.3, random_state=0)
#Learning
model1.fit(train_features3,train_labels3)
#The best hyper parameters set
print("Best Hyper Parameters:\n",model1.best_params_)
#Prediction
prediction3=model1.predict(test_features3)
#importing the metrics module
from sklearn import metrics
#evaluation(Accuracy)
print("Accuracy:",metrics.accuracy_score(prediction3,test_labels3))
#evaluation(Confusion Metrix)
print("Confusion Metrix:\n",metrics.confusion_matrix(prediction3,test_labels3))
pkl.dump(model1.best_params_,open('SVM_skipgram_param.pkl','wb'))
pkl.dump(model1.best_estimator_,open('SVM_skimgram_clf.pkl','wb'))
```
