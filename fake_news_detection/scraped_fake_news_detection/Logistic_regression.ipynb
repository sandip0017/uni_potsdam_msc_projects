# Predictive Analysis and Hyperparameter Tuning

The prediction of fake news mail is a classification problem. Therefore, we can use some popular Machine Learning Classifiers to make the prediction and then compare their results. Also, we are going to use a Neural Network to learn the Word Embeddings from our Model and measure its performance.

Our task at first is to detect the best Hyperparameters for the classifiers, analysze their performance on our dataset, and then select the best Predictive Model. We are going to use **Grid Search** to detect the best Hyperparameters for the classifiers. The classifiers that can be used for this comparison purpose are mentioned below

1. *Decision Tree*
2. *Random Forest*
3. *Logistic Regression*
4. *Multinomial Naive Bayes*
5. *Support Vector Machines with Gaussian Kernel*

```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
from scipy import sparse
from sklearn.model_selection import train_test_split
import pickle as pkl
```

```python
feature_mtx = np.load('model-features/feature_matrix.npy')
X_tfidf_sparse = sparse.load_npz('model-features/tf-idf-matrix.npz')
X_word_embed = np.load('model-features/wordembedding_feature.npy')

X_tfidf = np.concatenate((feature_mtx, X_tfidf_sparse.toarray()), axis=1)
X_embedding = np.concatenate((feature_mtx, X_word_embed), axis=1)

label = np.load('model-features/label.npy')
```

```python
#Hyperparameter tuning
#Logistic regression with TF-IDF
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
grid={"C":np.linspace(50,500,50), "penalty":["l1","l2"]}# l1 lasso l2 ridge
train_features4, test_features4, train_labels4, test_labels4 = train_test_split(X_tfidf, label, test_size=0.3, random_state=0)
logreg=LogisticRegression()
logreg_cv=GridSearchCV(logreg,grid,cv=3)
logreg_cv.fit(train_features4,train_labels4)
best_paraeter=logreg_cv.best_params_
Optimised_LR_TFIDF=logreg_cv.best_estimator_
print("tuned hpyerparameters :(best parameters) ",logreg_cv.best_params_)
```

```python
pkl.dump(best_paraeter,open('LR_TF-ID_param.pkl','wb'))
pkl.dump(Optimised_LR_TFIDF,open('LR_TF-ID_clf.pkl','wb'))
```

```python
logreg2=LogisticRegression(C=50,penalty="l2")# put the parameter
logreg2.fit(train_features4,train_labels4)
print("score",logreg2.score(test_features4,test_labels4))
```

```python
#Hyperparameter tuning
#Logistic regression with skipgram
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
grid={"C":np.linspace(50,500,50), "penalty":["l1","l2"]}# l1 lasso l2 ridge
train_features5, test_features5, train_labels5, test_labels5 = train_test_split(X_embedding, label, test_size=0.3, random_state=0)
logreg=LogisticRegression()
logreg_cv=GridSearchCV(logreg,grid,cv=3)
logreg_cv.fit(train_features5,train_labels5)
best_paraeter=logreg_cv.best_params_
Optimised_LR_TFIDF=logreg_cv.best_estimator_
print("tuned hpyerparameters :(best parameters) ",logreg_cv.best_params_)
```

```python
logreg2=LogisticRegression(C=2,penalty="l2")# put the parameter
logreg2.fit(train_features5,train_labels5)
print("score",logreg2.score(test_features5,test_labels5))
```

```python
pkl.dump(best_paraeter,open('LR_skipgram_param.pkl','wb'))
pkl.dump(best_paraeter,open('LR_skipgram_clf.pkl','wb'))
```
