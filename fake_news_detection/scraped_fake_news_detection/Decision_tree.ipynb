# Predictive Analysis and Hyperparameter Tuning

The prediction of fake news mail is a classification problem. Therefore, we can use some popular Machine Learning Classifiers to make the prediction and then compare their results. Also, we are going to use a Neural Network to learn the Word Embeddings from our Model and measure its performance.

Our task at first is to detect the best Hyperparameters for the classifiers, analysze their performance on our dataset, and then select the best Predictive Model. We are going to use **Grid Search** to detect the best Hyperparameters for the classifiers. The classifiers that can be used for this comparison purpose are mentioned below

1. *Decision Tree*
2. *Random Forest*
3. *Logistic Regression*
4. *Multinomial Naive Bayes*
5. *Support Vector Machines with Gaussian Kernel*

```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
from scipy import sparse
from sklearn.model_selection import train_test_split
import pickle as pkl
```

```python
feature_mtx = np.load('model-features/feature_matrix.npy')
X_tfidf_sparse = sparse.load_npz('model-features/tf-idf-matrix.npz')
X_word_embed = np.load('model-features/wordembedding_feature.npy')

X_tfidf = np.concatenate((feature_mtx, X_tfidf_sparse.toarray()), axis=1)
X_embedding = np.concatenate((feature_mtx, X_word_embed), axis=1)

label = np.load('model-features/label.npy')
```

```python
#Hyper Parameters Tuning
#DesicionTree with TF-IDF
#importing modules
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier
#making the instance
model= DecisionTreeClassifier(random_state=0, max_depth = 2)
#Hyper Parameters Set
params = {'max_features': ['auto', 'sqrt', 'log2'],
          'min_samples_split': np.arange(2,15), 
          'min_samples_leaf':np.arange(1,12),
          'random_state':[0]}
#Making models with hyper parameters sets
model1 = GridSearchCV(model, param_grid=params, n_jobs=4)
#Learning
train_features, test_features, train_labels, test_labels = train_test_split(X_tfidf, label, test_size=0.3, random_state=0)
model1.fit(train_features,train_labels)
#The best hyper parameters set
print("Best Hyper Parameters:",model1.best_params_)
optimised_DT_TFIDF = model1.best_estimator_
```

```python
pkl.dump(model1.best_params_, open('decision_tree_TF-ID_param.pkl', 'wb'))
pkl.dump(model1.best_estimator_, open('decision_tree_TF-ID_clf.pkl', 'wb'))
```

```python
#Hyper Parameters Tuning
#DesicionTree with Skipgram
#importing modules
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier
#making the instance
model= DecisionTreeClassifier(random_state=0, max_depth = 2)
#Hyper Parameters Set
params = {'max_features': ['auto', 'sqrt', 'log2'],
          'min_samples_split': np.arange(2,15), 
          'min_samples_leaf':np.arange(1,12),
          'random_state':[0]}
#Making models with hyper parameters sets
model1 = GridSearchCV(model, param_grid=params, n_jobs=4)
#Learning
train_features1, test_features1, train_labels1, test_labels1 = train_test_split(X_embedding, label, test_size=0.3, random_state=0)
model1.fit(train_features1,train_labels1)
#The best hyper parameters set
print("Best Hyper Parameters:",model1.best_params_)
#Prediction
prediction1=model1.predict(test_features1)
#importing the metrics module
from sklearn import metrics
#evaluation(Accuracy)
print("Accuracy:",metrics.accuracy_score(prediction1,test_labels1))
#evaluation(Confusion Metrix)
print("Confusion Metrix:\n",metrics.confusion_matrix(prediction1,test_labels1))
```

```python
pkl.dump(model1.best_params_, open('decision_tree_sg_param.pkl', 'wb'))
pkl.dump(model1.best_estimator_, open('decision_tree_sg_clf.pkl', 'wb'))
```
